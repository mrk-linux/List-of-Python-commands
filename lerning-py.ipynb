```python
def shift_alphabet(word):
    result = ""
    for char in word:
        if char.isalpha():
            shifted_char = chr((ord(char.lower()) - ord('a') + 1) % 26 + ord('a'))
            result += shifted_char
        else:
            result += char
    return result

# مثال:
input_word = input("لطفاً یک کلمه وارد کنید: ")
shifted_word = shift_alphabet(input_word)
print("نتیجه:", shifted_word)
```

```python
#بدست اوردن فاکتوریل
def fact(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f

def s_fact(n):
    s = 0
    for i in range(1, n+1):
        s += fact(i)
    return s

number = int(input("enter a number:"))
print("s:", s_fact(number))
```

```python
# حل مثال برای دکوراتور

passwords = {"ali":"123", "reza":"456", "alireza":"789"}
blacklist = {"ali"}
from functools import wraps

def ban(func):
    @wraps(func)
    def inner(*args, **kwargs):
        if args[0] in blacklist:
            print("This user is blocked!")
        else:
            func(*args, **kwargs)
    return inner

@ban 
def print_passwords(username):
    print(username, ":", passwords[username])

def change_password(username, new_password):
    passwords[username] = new_password
    print(username, ":", passwords[username])

print_passwords("reza")
change_password("alireza","new password")
```

```python
# حل مثال برای دکوراتور

from time import perf_counter
from functools import wraps

def time_calculation(func):
    @wraps(func)
    def warpper_decorator(*args, **kwargs):
        start_time = perf_counter()
        value = func(*args, **kwargs)
        end_time = perf_counter()
        run_time = end_time - start_time
        print("the run time fo", func.__name__, "is", run_time)
        return value
    return warpper_decorator

@time_calculation
def A(y):
    s = 0
    for i in range(y):
        s += i**2

@time_calculation
def B(x):
    fact = 1
    for i in range(1, x+1):
        fact *= i


A(10000)
B(100)
```

```python
# حل مثال برای ژنراتور
def list_range(start, end, step=1):
    new_rang = []
    while start < end :
        new_rang.append(start)
        start += step
    return new_rang

r = range(10, 20, 2)
print(list(r))
lr = list_range(10, 20, 2)
print(list(lr))

# حالا با استفاده از ژنراتور
def gen_range(start, end, step=1):
    while start < end:
        yield start
        start += step

lr = list_range(10, 20, 2)
print(list(lr))

gr = gen_range(10, 20, 2)
print(list(gr))
```

```python
# برنامه ای بنویسید که محتوا ی فایلی به نام ”txt.sample ”را 
# بخواند و تکرار هر کلمه را بشمارد و سپس کلمه و تعداد آن را چاپ 
# کنی د.

import re

f = open("text.text", "r", encoding="utf-8")
y = f.read()
def count_words(y):
    # پیدا کردن همه کلمات با حروف فارسی/انگلیسی
    words = re.findall(r'[\u0600-\u06FFa-zA-Z]+', y)

    word_count = {}
    for word in words:
        lower_word = word.lower()
        if lower_word in word_count:
            word_count[lower_word] += 1
        else:
            word_count[lower_word] = 1
    return word_count


counts = count_words(y)
for word, count in counts.items():
    print(f"'{word}': {count}")

 
#2
import re
from collections import defaultdict

f = open("text.text", "r", encoding="utf-8")
y = f.read()
def count_words(y):
    # پیدا کردن همه کلمات با حروف فارسی/انگلیسی
    words = re.findall(r'[\u0600-\u06FFa-zA-Z]+', y)

    word_count = defaultdict(int)
    for word in words:
        lower_word = word.lower()
        word_count[lower_word] += 1
    return word_count

counts = count_words(y)
for word, count in counts.items():
    print(f"'{word}': {count}")
```

```python
def factorial(n: int) -> int:
    """
    محاسبه فاکتوریل یک عدد
    
    Args:
        n (int): عدد ورودی (باید >= 0 باشد)
        
    Returns:
        int: نتیجه فاکتوریل
        
    Examples:
        >>> factorial(5)
        120
        >>> factorial(0)
        1
        >>> factorial(-1)
        Traceback (most recent call last):
            ...
        ValueError: n must be >= 0
    """
    if n < 0:
        raise ValueError("n must be >= 0")
    return 1 if n <= 1 else n * factorial(n - 1)

if __name__ == "__main__":
    import doctest
    doctest.testmod()
```

 کلاس باید دارای اتربیوت های طول عرض و ارتفا  متد های مساحت ومحیط باشد


```python
class Rectangle:

    def __init__(self, length, width, height)-> None:
        self.length = length
        self.width = width
        self.height = height

    def Area(self):
        """محاسبه حجم مکعب مستطیل"""
        return self.length * self.width * self.height

    def Perimeter(self):
        """محاسبه مجموع یال‌های مکعب مستطیل"""
        return 4 * (self.length + self.width + self.height)
        
    def surface_area(self):
        """محاسبه مساحت سطح (اختیاری)"""
        return 2 * (self.length * self.width + 
                   self.length * self.height + 
                   self.width * self.height)
    
    def __str__(self):
        return (f"Rectangle: length={self.length}, width={self.width}, height={self.height}\n"
                f"area: {self.Area()}\n"
                f"Perimeter: {self.Perimeter()}\n"
                f"surface area: {self.surface_area()}")

Rectangle1 = Rectangle(23,  21,  20)
print(Rectangle1)
```

## mixin

1. یک (mixin) به نام(PrintinfoMixin) 

ایجاد کنید که متدی با نام (print_ino) را به یک کلاس اضافه کند

متد (print_ino) باید نام کلاس و ویژگی های ان را چاپ کند

```python

class PrintinfoMixin:
    def print_info(self):
        print(f"Class name: {self.__class__.__name__}\n Attributes: {self.__dict__}")

class Conditions:
    def move(self):
        print("animal moving for 3km")

class animal(Conditions, PrintinfoMixin):
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __str__(self):
        return (f"Animal Name: {self.name}\nAge: {self.age}")

name = input("enter animal name")
age = input("enter animal age")
animal1 = animal(name, age)
#print(animal)
animal1.print_info()
animal1.move()
```

```python
class PrintInfoMixin:
    def print_info(self):
        # چاپ نام کلاس
        print(f"Class name: {self.__class__.__name__}")
        # چاپ همه ویژگی‌های شیء
        print("Attributes:")
        for key, value in self.__dict__.items():
            print(f"  {key}: {value}")

# کلاس پایه
class Conditions:
    def move(self):
        print("animal moving for 3km")

# کلاس Animal که از Conditions و Mixin ارث‌بری می‌کند
class Animal(Conditions, PrintInfoMixin):
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Animal Name: {self.name}\nAge: {self.age}"

# تست
name = input("enter animal name")
age = input("enter animal age")
animal1 = Animal(name, age)
animal1.print_info()
animal1.move()
# lin 10 , 29 (me) & line 4 , 8 (ai)
```

یک mixin به نام MathOperationsMixin ایجاد کنید

که عملیات ریاضی پایه(جمع ,تفریق ,ضرب و تقسیم ) را به یک کلاس اضافه کند

کلاس مورد نظر باید دارای اتربیوت های x وy 

```python
class MathOperationsMixin():
    def plas(self):
        z = (self.x) + self.y
        print("plas z=", z)

    def negative(self):
        z = self.x - self.y
        print("negative z=", z)

    def multiply(self):
        z = self.x * self.y
        print("multiply z=", z)

    def divide(self):
        z = self.x / self.y
        print("divide z=", z)
class Calculator(MathOperationsMixin):
    def __init__(self, x: int, y: int):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Calculator with x={self.x}, y={self.y}"

#ورودی‌های input همیشه به صورت رشته (string) ذخیره می‌شوند.
#  وقتی آن‌ها را مستقیم به کلاس می‌دهی، self.x و self.y رشته هستند و عملیات ریاضی مثل جمع یا ضرب روی رشته‌ها خطا یا رفتار غیرمنتظره ایجاد می‌کند.
x1 = int(input("enter number of the x: "))
y1 = int(input("enter number of the y: "))

Calculator1 = Calculator(x1, y1)
print(Calculator1)  
Calculator1.plas()
Calculator1.negative()
Calculator1.multiply()
Calculator1.divide()
```

## مثال کامنت

```python
from datetime import datetime
from time import sleep

class Product:
    def __init__(self, product_name, price, off):
        self.product_name = product_name
        self.price = price
        self.off = off

    def __str__(self):
        return self.product_name


class Comment:
    website_name = "alimrk.ir"


    #instance method
    def __init__(self, product, name, description, like, dislike):
        self.product = product
        self.name = name
        self.description = description
        self.like = like
        self.dislike = dislike
        self.date = datetime.now()

    #instance method
    def show(self):
        print(
            f"product: {self.product}\n"
            f"description: {self.description}\n"
            f"name: {self.name}\n"
            f"date: {self.date}\n"
            f"like: {self.like} | dislike: {self.dislike}"
        )

    @classmethod
    def info(cls):
        print(f"website name: {cls.website_name}")

    @classmethod
    def censorship(cls, product, name, description, like, dislike):
        print("the comment was censored!!!")
        cs = description.replace("بیشعور", "****")
        return cls(product, name, cs, like, dislike)

    @staticmethod
    def elapsed_time(start_time):
        sleep(3)
        now = datetime.now()
        print("elapsed seconds:", (now - start_time).seconds)


python_course = Product("python expert", 0, 0)

c1 = Comment.censorship(python_course, "alireza", "دوره بیشعوری بود", 15, 76)
c1.show()

print("-" * 40)
Comment.info()

print("-" * 40)
c2 = Comment(python_course, "alireza", "دوره خوبی بود", 17, 6)
c2.show()

print("-" * 40)
Comment.elapsed_time(c2.date)
```

## ستر، گتر و پروپرتی

```python
class Color:
    def __init__(self, rgb, name):
        # استفاده از نام عمومی (بدون آندرکور) باعث می‌شود حتی در لحظه ساخت آبجکت،
        # متدهای ستِر (Setter) صدا زده شوند و اعتبارسنجی انجام شود [۲۱، ۲۵، ۲۷].
        self.rgb = rgb
        self.name = name

    # --- متدهای مربوط به ویژگی name ---
    def _get_name(self):
        # استفاده از آندرکور برای جلوگیری از حلقه بی‌نهایت [۲۸، ۲۹]
        return self._name

    def _set_name(self, name):
        # اعتبارسنجی: اگر نام خالی باشد، خطا صادر می‌شود [۱۰، ۱۱]
        if name:
            self._name = name
        else:
            # استفاده از رِپر (!r) برای نمایش دقیق کوتیشن‌ها در پیام خطا [۱۱، ۱۲]
            raise ValueError(f"Invalid Name! {name!r}")

    # --- متدهای مربوط به ویژگی rgb ---
    def _get_rgb(self):
        return self._rgb

    def _set_rgb(self, rgb):
        # در اینجا نیز می‌توان اعتبارسنجی‌های مربوط به کد رنگ را اضافه کرد [۲۱]
        self._rgb = rgb

    # تعریف پروپرتی‌ها برای ایجاد رابط کاربری ساده و تمیز [۱۹، ۲۱، ۲۲]
    name = property(_get_name, _set_name)
    rgb = property(_get_rgb, _set_rgb)

# --- تست برنامه ---
try:
    # ساخت آبجکت با استفاده از کد هگزادسیمال (مبنای ۱۶) [۲، ۳]
    c1 = Color(0xadd8e6, "Light Blue")
    print(f"Name: {c1.name}, RGB: {c1.rgb}")

    # تغییر مقدار (پایتون خودکار متد سِتر را صدا می‌زند) [۲۳، ۲۹]
    c1.name = "Blue"
    print(f"Updated Name: {c1.name}")

    # تست اعتبارسنجی با ارسال رشته خالی [۱۱، ۲۰]
    c1.name = "" 
except ValueError as e:
    print(e)
```

یک کلاس (circle) با یک اتربیوت (radius) ایجاد کنید و یک (property) برای محاسبه مساحت بنویسید

```python
class circle:
    pi = 3.14

    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        return self.pi * self.radius * self.radius

    @property
    def perimeter(self):
        return 2 * self.pi * self.radius

circle1 = circle(5)
print("area:", circle1.area)
print("perimeter:", circle1.perimeter)
```

کلاس (circle) تمرین قبلی را گسترش دهید و یک متد (setter) برای ویژگی (radius) اضافه کنید.

متد (setter) باید اطمینان حاصل کند که شعاع همیشه مثبت است

```python
class Circle:
    pi = 3.14

    def __init__(self, radius):
        self.radius = radius  

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Radius must be non-negative")
        self._radius = value

    @property
    def area(self):
        return self.pi * self.radius * self.radius
    
    @property
    def perimeter(self):
        return 2 * self.pi * self.radius


circle1 = Circle(5)
print("area:", circle1.area)
print("perimeter:", circle1.perimeter)

circle1.radius = -10   # اینجا خطا می‌دهد
```
